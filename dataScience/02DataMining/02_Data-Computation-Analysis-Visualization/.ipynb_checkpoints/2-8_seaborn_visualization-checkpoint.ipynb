# Seaborn与可视化
Seaborn是斯坦福大学出的一个非常好用的可视化包，这一节我们一起来看看这个包和相关的一些用法。

```python
%matplotlib inline
%pylab inline
%config ZMQInteractiveShell.ast_node_interactivity='all'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make the graphs a bit prettier, and bigger
sns.set_style('darkgrid')
pd.set_option('display.mpl_style', 'default')
# matplotlib.style.use('ggplot')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.family'] = 'sans-serif'

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
```

#### 先用pandas读进来一份数据

```python
names = [
       'mpg'
    ,  'cylinders'
    ,  'displacement'
    ,  'horsepower'
    ,  'weight'
    ,  'acceleration'
    ,  'model_year'
    ,  'origin'
    ,  'car_name'
]
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data", sep='\s+', names=names)
df['maker'] = df.car_name.map(lambda x: x.split()[0])
df.origin = df.origin.map({1: 'America', 2: 'Europe', 3: 'Asia'})
df=df.applymap(lambda x: np.nan if x == '?' else x).dropna()
df['horsepower'] = df.horsepower.astype(float)
df.head()
```

# 一般绘图：factorplot 和 FacetGrid

### 8.1.1 根据2个维度变量绘图

```python
df_bak=df.copy()
df_bak.info()
```

```python
# 画出model_year和mpg的关系图
# 圆点：mean 平均数
sns.factorplot(data=df, x="model_year", y="mpg")
# df[['model_year','mpg']].groupby('model_year').aggregate('mean').plot()
# df[['model_year','mpg']].groupby('model_year').aggregate('median').plot()
# df[['model_year','mpg']].groupby('model_year').aggregate('sum').plot()
```

### 8.1.2 可以按照第3个维度绘制不同的关系图

```python
sns.factorplot(data=df, x="model_year", y="mpg", col="origin")
# df[['origin','model_year','mpg']].groupby(['origin','model_year']).aggregate('mean')
sns.factorplot(data=df, x="model_year", y="mpg", col="origin", kind='bar')
```

### 8.1.3 可以从折线图切成柱状图

```python
sns.factorplot("model_year", data=df, col="origin", kind='count')
# df1=df[['model_year','origin']]
# df1['freq']=1
# df1.groupby(['origin','model_year']).aggregate('count')
```

```python
df=df_bak
df_mpg_aggr=df[['mpg','origin']].groupby('origin').aggregate('sum').reset_index()
df_mpg_aggr.columns=['origin','mpg_aggr']
df=df.merge(df_mpg_aggr)
df['mpg_prob']=df['mpg']/df['mpg_aggr']
df['mpg_int']=df['mpg'].apply(int)
df_new=df[['origin','mpg_prob','mpg_int']].groupby(['origin','mpg_int']).aggregate('sum').reset_index()
sns.factorplot(data=df_new, x="mpg_int", y="mpg_prob", col="origin")
```

```python
sns.FacetGrid(df, col="origin")
```

```python
sns.FacetGrid(df, col="origin").map(sns.distplot, "mpg")

sns.FacetGrid(df).map(sns.distplot, "mpg")
```

### 8.1.4 散点图

```python
g = sns.FacetGrid(df, col="origin", hue='origin')
g.map(plt.scatter, "horsepower", "mpg", alpha=.3)
plt.xlim(0, 250)
plt.ylim(0, 60)
```

### 8.1.5 绘图的同时还做回归

```python
g = sns.FacetGrid(df, col="origin", hue='origin')
g.map(sns.regplot, "horsepower", "mpg")
plt.xlim(0, 250)
plt.ylim(0, 60)
```

### 8.1.6 kde等高线图

```python
df['tons'] = (df.weight/2000).astype(int)
g = sns.FacetGrid(df, col="origin", row="tons", hue='origin')
g.map(sns.kdeplot, "horsepower", "mpg")
plt.xlim(0, 250)
plt.ylim(0, 60)
```

### 8.1.7 按照2个维度展开画图

```python
g = sns.FacetGrid(df, col="origin", row="tons")
g.map(plt.hist, "mpg", bins=np.linspace(0, 50, 11))
plt.xlim(0, 50)
plt.ylim(0, 70)
```

# pairplot and PairGrid

### 8.2.1 多个维度两两组合绘图

```python
g = sns.pairplot(df[["mpg", "horsepower", "weight", "origin"]], hue="origin",
                 kind='scatter', diag_kind="hist", size=5)
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)
```

### 8.2.2 组合绘图的时候顺便回归一下

```python
g = sns.PairGrid(df[["mpg", "horsepower", "weight", "origin"]], hue="origin")
g.map_upper(sns.regplot)
g.map_lower(sns.residplot)
g.map_diag(plt.hist)
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)
g.add_legend()
g.set(alpha=0.01)
```

# jointplot and JointGrid

## 联合绘图(kde等高)

```python
sns.jointplot("mpg", "horsepower", data=df, kind='kde')

sns.FacetGrid(df, size=1.5, aspect=3.5).map(sns.distplot, "mpg")
plt.xlim(0, 50)
plt.ylim(0, 0.05)
sns.FacetGrid(df, size=1.5, aspect=3.5).map(sns.distplot, "horsepower")
plt.xlim(0, 250)
plt.ylim(0, 0.0175)
sns.FacetGrid(df, size=5).map(sns.kdeplot, "mpg", "horsepower")
plt.xlim(0, 50)
plt.ylim(0, 250)
```

## 联合绘图(加回归)

```python
sns.jointplot("mpg", "horsepower", data=df, kind='reg', order=2)
```

## 联合绘图(自定义绘图参数)

```python
g = sns.JointGrid(x="mpg", y="horsepower", data=df)
g.plot_joint(sns.regplot, order=2)
g.plot_marginals(sns.distplot)
```

# 总结

```python
%matplotlib inline
%pylab inline
%config ZMQInteractiveShell.ast_node_interactivity='all'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make the graphs a bit prettier, and bigger
sns.set_style('darkgrid')
pd.set_option('display.mpl_style', 'default')
# matplotlib.style.use('ggplot')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.family'] = 'sans-serif'

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

names = [
       'mpg'
    ,  'cylinders'
    ,  'displacement'
    ,  'horsepower'
    ,  'weight'
    ,  'acceleration'
    ,  'model_year'
    ,  'origin'
    ,  'car_name'
]
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data", sep='\s+', names=names)
df['maker'] = df.car_name.map(lambda x: x.split()[0])
df.origin = df.origin.map({1: 'America', 2: 'Europe', 3: 'Asia'})
df=df.applymap(lambda x: np.nan if x == '?' else x).dropna()
df['horsepower'] = df.horsepower.astype(float)

##factorplot
# 折线图，统计 两个变量间的关系，y取 平均数mean
sns.factorplot(data=df, x="model_year", y="mpg")
sns.factorplot(data=df, x="model_year", y="mpg", col="origin")
# 切成 柱状图，y取 平均数mean
sns.factorplot(data=df, x="model_year", y="mpg", col="origin", kind='bar')

# 柱状图，统计 单变量model_year 频率
sns.factorplot("model_year", data=df, col="origin", kind='count')
# df1=df[['model_year','origin']]
# df1['freq']=1
# df1.groupby(['origin','model_year']).aggregate('count')


## FacetGrid
# 概率分布直方图，并拟合 连续性概率密度曲线
g=sns.FacetGrid(df, col="origin")
g.map(sns.distplot, "mpg")
sns.FacetGrid(df).map(sns.distplot, "mpg")
# 散点图
g = sns.FacetGrid(df, col="origin", hue='origin')
g.map(plt.scatter, "horsepower", "mpg", alpha=.3)
plt.xlim(0, 250)
plt.ylim(0, 60)

# 散点图，并做回归
g = sns.FacetGrid(df, col="origin", hue='origin')
g.map(sns.regplot, "horsepower", "mpg")
plt.xlim(0, 250)
plt.ylim(0, 60)

df['tons'] = (df.weight/2000).astype(int)
# 散点图（多元图形）
g = sns.FacetGrid(df, col="origin", row="tons", hue='origin')
g.map(plt.scatter, "horsepower", "mpg", alpha=.3)
plt.xlim(0, 250)
plt.ylim(0, 60)

# kde等高线图，圈圈表示数据聚集的区域
# 圈越小，数据越多
g = sns.FacetGrid(df, col="origin", row="tons", hue='origin')
g.map(sns.kdeplot, "horsepower", "mpg")
plt.xlim(0, 250)
plt.ylim(0, 60)

# 概率分布直方图，并拟合 连续性概率密度曲线
g = sns.FacetGrid(df, col="origin", row="tons")
g.map(sns.distplot, "mpg", bins=np.linspace(0, 50, 11))
plt.xlim(0, 50)
plt.ylim(0, 0.25)

# 频率分布直方图
g=sns.FacetGrid(df, col="origin", row="tons")
g.map(plt.hist, "mpg", bins=np.linspace(0, 50, 11))
plt.xlim(0, 50)
plt.ylim(0, 70)
# 同上，但要自己计算 分布的间隔
df['mpg_5']=(df.mpg/5).astype(int)
sns.factorplot("mpg_5", data=df, col="origin", row="tons", kind='count')
plt.xlim(0, 10)
plt.ylim(0, 70)


##pairplot
# 多个维度 ["mpg", "horsepower", "weight"] 两两组合绘图
g = sns.pairplot(df[["mpg", "horsepower", "weight", "origin"]], hue="origin",
                 kind='scatter', diag_kind="hist", size=5)
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)
    

##PairGrid
# 组合绘图，并回归
g = sns.PairGrid(df[["mpg", "horsepower", "weight", "origin"]], hue="origin")
g.map_upper(sns.regplot)
g.map_lower(sns.residplot)
g.map_diag(plt.hist)
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)
g.add_legend()
g.set(alpha=0.01)


##jointplot
# 多个变量联合绘图，一次性搞定
sns.jointplot("mpg", "horsepower", data=df, kind='kde')
# 拆分看看
sns.FacetGrid(df, size=1.5, aspect=3.5).map(sns.distplot, "mpg")
plt.xlim(0, 50)
plt.ylim(0, 0.05)
sns.FacetGrid(df, size=1.5, aspect=3.5).map(sns.distplot, "horsepower")
plt.xlim(0, 250)
plt.ylim(0, 0.0175)
sns.FacetGrid(df, size=5).map(sns.kdeplot, "mpg", "horsepower")
plt.xlim(0, 50)
plt.ylim(0, 250)
# 加回归参数
sns.jointplot("mpg", "horsepower", data=df, kind='reg', order=2)


##JointGrid
# 联合绘图：自定义绘图参数
g = sns.JointGrid(x="mpg", y="horsepower", data=df)
g.plot_joint(sns.regplot, order=2)
g.plot_marginals(sns.distplot)
```
