### Python解释与运行


Python解释器interpreter
>运行在系统线程上的C语言函数+超大循环。每个字节码指令均对应一个完全由C实现的逻辑。


Python解释器在执行过程
>先对文件中的Python源代码进行编译，编译的主要结果是产生的一组Python的字节码（byte code），pyc文件是字节码在磁盘上的表现形式，是进行import module时对py文件进行编译的文件结果（如果之前编译过，且源文件又没有变化，则直接运行该pyc文件），而编译结果就是Python中的PyCodeObject对象，然后PyCodeObject对象交给Python虚拟机（Virtual Machine），由虚拟机按照顺序一条一条地执行字节码，从而完成对Python程序的执行动作。


> python： .py  -->(编译器)-->.pyc-->解释器(虚拟机)-->返回结果   

> 在编译好的Python环境中，编译器和解释器都被封装进了python36.dll中，由python.exe调用

### 三种赋值方式 = copy deepcopy *** 重要！！！ ***

```python
import copy
a=[1,2,3,[4,5,6]]
print("引用传递 指向同一内存地址")
b=a#赋值
a,b
print("change a")
a[3].append(777)
a,b
print("change a")
a.append(7)
a,b
print("change b")
b[3].append(77777)
a,b
print("change b")
b.append(9)
a,b

print("shallow copy 最外层互不影响")
a=[1,2,3,[4,5,6]]
c=copy.copy(a)#shallow copy
a,c
print("change a")
a[3].append(17)
a,c
print("change a")
a.append(11)
a,c
print("change c")
c[3].append(27)
a,c
print("change c")
c.append(22)
a,c

print("deep copy 值传递 都互不影响")
a=[1,2,3,[4,5,6]]
d=copy.deepcopy(a)#Deep copy
a,d
print("change a")
a[3].append(117)
a,d
print("change a")
a.append(111)
a,d
print("change c")
d[3].append(227)
a,d
print("change c")
d.append(222)
a,d
```

### 可变对象(Array, List, Tuple, Dict等)

```python
#Array[]:相同类型元素的序列
from array import array
a1=array('h', [1,2,3])
#memoryview属于引用，不产生新的内存
memv=memoryview(a1)#内存视图
print("memoryview: {}, the lengeth of memoryview: {}".format(memv.tolist(), len(memv)))
memv[0]=55
print("*** changed memory view with memv[0]=55")
print("old array: {}, memoryview: {}\n".format(a1.tolist(), memv.tolist()))


#复制操作：对bytearray对象片断是复制，而非直接引用。产生新的内存
str1=bytearray('中国','utf-8')
print("Bytearray Object: {}".format(str1))#序列访问
tempstr=str1[0:3]#片断赋值给新变量
tempstr[0:3]=b'\xff\xff\xff'
print("*** changed bytearray with bytearray[0:3]={} *** No Change!".format(tempstr))
print("old bytearray: {}\n".format(str1))#原str1没有变化，因此是发生了复制。。


#直接操作
#bytes,bytearray,array.arrray支持视图操作
v=memoryview(str1)#对str1进行内存视图
v[0:3]=b'\xff\xff\xff'#此时对视图对象操作，即是对原str1操作了
print("*** changed bytearray with memoryview[0:3]={} *** Changed!".format(v.tobytes()))
print("old bytearray: {}".format(str1))
```

#### 延伸阅读：追踪(traceback) 

当简单的异常处理给我们的信息已经无法解决调试问题时，可以使用traceback对象进行深入观察。

追踪是在出现异常时用于回溯的对象，与栈帧相反。由于异常时才会构建，而异常未捕获时会一直向外层栈帧抛出，所以需要使用try才能见到这个对象。   
你可以使用sys模块的exc_info()函数获得它，这个函数返回一个元组，元素分别是异常类型、异常对象、追踪。   
对象中包含出错的行数、位置等数据。


traceback的属性全部是只读，例如如下属性:
- tb_next: 追踪的下一个追踪对象
- tb_frame: 当前追踪对应的栈帧
- tb_lineno: 当前追踪的行号

```python
import sys, traceback
    
try:
    a=never_named_value
    print(int(a))
except ValueError:#发生ValueError错误执行以下代码
    print ("Error: 函数参数传值错误")
except NameError:#发生NameError错误执行以下代码
    print ("Error: 名称空间使用出错")
    
    #使用sys.exc_info获得tb对象
    exc_type, exc_value, exc_traceback_obj=sys.exc_info()
    print('元组内三个对象: {}, {}, {}'.format(exc_type, exc_value, exc_traceback_obj))
    
    #直接打印tb对象内的主要信息
    traceback.print_exc() 
    #traceback对象内更多内容
    print('tb对象内更多信息: {}, {}'.format(exc_traceback_obj.tb_frame, exc_traceback_obj.tb_lasti))
else:
    print ("Error: others")
finally:
    print('Everything is fine')
    
print('Try except Over!')
    
#try-finally 无法是否捕获了异常，都会执行finally后的语句
```

### 装饰器：在不改动签名的前提下，为一个或多个函数同时增加部分功能

```python
# 实现装饰器
def printFunction(func):
    print "Call the function {}".format(func.__code__.co_name)
    return func

@printFunction
def getSum(*argv):
    return sum(argv)

@printFunction
def getMax(*argv):
    return max(argv)

print getSum(1,2,3,4)
print getMax(2,3)
```

```python
# 增添计时器功能
import datetime

def getRunTime(fun):
    def inner():
        print("exec the function")
        start=datetime.datetime.now()
        fun()
        end=datetime.datetime.now()
        print('Run time:',end-start)
      
    return inner

#装饰器语法糖在要被包裹的函数前声明。
#@后面的函数名，是把下边函数进行包裹的函数的名字outer_fun
@getRunTime#为fun1函数添加一个装饰器函数叫outer_fun
def fun():
    return 'this is fun function--'

fun()
```

## 面向对象
* 类变量、实例变量
* 类方法、实例方法
* 继承
* 多态

```python
class Cinema():
    #类变量
    cinema_cnt=0
    sales_cnt=0
    
    #类方法：初始化方法，并初始相应的类变量实例变量
    def __init__(self,name,location,sales):
        #实例变量
        self.name=name
        self.location=location
        self.sales=sales
        Cinema.cinema_cnt+=1
        Cinema.sales_cnt+=sales
    
    #类方法：获取全部电影院实际销售的方法
    @classmethod
    def getSales(cls,):
        return Cinema.sales_cnt
    
    #实例方法：要求更新实例的销售总额及类的销售总额
    def saleTickets(self,sales):
        self.sales+=sales
        Cinema.sales_cnt+=sales

#继承
class MiniCinema(Cinema):
    #多态
    def saleTickets(self,sales):
        if sales>100:
            sales*=0.9
        super().saleTickets(sales)

carl=MiniCinema('carl','Beijing',200)
print("init a miniCinema:")
carl.getSales()

carl.saleTickets(200)
print("saled tickets with 200:")
carl.getSales()
```

### 正则表达式
* [正则表达式基础知识](https://deerchao.net/tutorials/regex/regex.htm)
* [re.match 从字符串的起始位置匹配一个模式](http://www.runoob.com/python/python-reg-expressions.html#flags)
* re.search 扫描整个字符串并返回第一个成功的匹配
* re.sub 用于替换字符串中的匹配项

```python
import re

# 简单匹配
print(re.match('www', 'www.baidu.com').span())  # 在起始位置匹配
print(re.match('com', 'www.baidu.com'))         # 不在起始位置匹配
# 多匹配: re.M: 多行匹配，影响 ^ 和 $; re.I: 使匹配对大小写不敏感
line = "Dog eat meat, hahaha"
matchObj = re.match( r'(.*) eat (.*?) (.*)', line, re.M|re.I)
if matchObj:
    print('matchObj.group() : ', matchObj.group())
    print('matchObj.group(1) : ', matchObj.group(1))
    print('matchObj.group(2) : ', matchObj.group(2))
    print('matchObj.group(3) : ', matchObj.group(3))
else:
    print('No match!!')

# 扫描整个字符串并返回第一个成功的匹配
print(re.search('com', 'www.runoob.com').span())  # 不在起始位置匹配

# 替换字符串中的匹配项
phone = "2004-959-559 # 这是一个国外电话号码"
re.sub(r'#.*$', '', phone) # 删除字符串中的 Python注释 
re.sub(r'\D', "", phone) # 删除非数字(-)的字符串 
re.sub(r'\d', "", phone) # 删除数字的字符串 

# 切分字符串
re.split(r'[\s\,\;]+', 'a,b;; c,;  d')

s = 'alice.jack-b@gmail.com'
match = re.search(r'^[\w\.\+\-]+\@[\w]+\.[a-z]{2,3}$', s)
if match:
    print(match.group())
```

## 数据处理/分析/可视化
### 简单数据处理流程：查看每个月的温度分布和下雪比例

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')
# matplotlib.style.use('ggplot')   #python3
plt.rcParams['figure.figsize'] = (15, 5)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False   #用来正常显示负号

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

# Load data
weather_2012 = pd.read_csv('./data/weather_2012.csv',
                           parse_dates=True,
                           index_col='Date/Time')

# Data Preprocessing
temperature = weather_2012['Temp (C)'].resample('M', how=np.median) #采样频率 1个月，统计 中位数
is_snowing = weather_2012['Weather'].str.contains('Snow')           #统计 字符串 的值是否包含某个字段，
snowiness = is_snowing.astype(float).resample('M', how=np.mean)     #采样频率 1个月，统计 平均值
# Rename the Dataframe
temperature.name = "Temperature"
snowiness.name = "Snowiness"
# Concat all of the Dataframe
stats = pd.concat([temperature, snowiness], axis=1)

# Plot
stats.plot(kind='bar', subplots=True, figsize=(15, 10))
```

### seaborn画图

```python
%matplotlib inline
%pylab inline
%config ZMQInteractiveShell.ast_node_interactivity='all'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make the graphs a bit prettier, and bigger
sns.set_style('darkgrid')
pd.set_option('display.mpl_style', 'default')
# matplotlib.style.use('ggplot')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.family'] = 'sans-serif'

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

names = [
       'mpg'
    ,  'cylinders'
    ,  'displacement'
    ,  'horsepower'
    ,  'weight'
    ,  'acceleration'
    ,  'model_year'
    ,  'origin'
    ,  'car_name'
]
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data", sep='\s+', names=names)
df['maker'] = df.car_name.map(lambda x: x.split()[0])
df.origin = df.origin.map({1: 'America', 2: 'Europe', 3: 'Asia'})
df=df.applymap(lambda x: np.nan if x == '?' else x).dropna()
df['horsepower'] = df.horsepower.astype(float)

##factorplot
# 折线图，统计 两个变量间的关系，y取 平均数mean
sns.factorplot(data=df, x="model_year", y="mpg")
sns.factorplot(data=df, x="model_year", y="mpg", col="origin")
# 切成 柱状图，y取 平均数mean
sns.factorplot(data=df, x="model_year", y="mpg", col="origin", kind='bar')

# 柱状图，统计 单变量model_year 频率
sns.factorplot("model_year", data=df, col="origin", kind='count')
# df1=df[['model_year','origin']]
# df1['freq']=1
# df1.groupby(['origin','model_year']).aggregate('count')


## FacetGrid
# 概率分布直方图，并拟合 连续性概率密度曲线
g=sns.FacetGrid(df, col="origin")
g.map(sns.distplot, "mpg")
sns.FacetGrid(df).map(sns.distplot, "mpg")
# 散点图
g = sns.FacetGrid(df, col="origin", hue='origin')
g.map(plt.scatter, "horsepower", "mpg", alpha=.3)
plt.xlim(0, 250)
plt.ylim(0, 60)

# 散点图，并做回归
g = sns.FacetGrid(df, col="origin", hue='origin')
g.map(sns.regplot, "horsepower", "mpg")
plt.xlim(0, 250)
plt.ylim(0, 60)

df['tons'] = (df.weight/2000).astype(int)
# 散点图（多元图形）
g = sns.FacetGrid(df, col="origin", row="tons", hue='origin')
g.map(plt.scatter, "horsepower", "mpg", alpha=.3)
plt.xlim(0, 250)
plt.ylim(0, 60)

# kde等高线图，圈圈表示数据聚集的区域
# 圈越小，数据越多
g = sns.FacetGrid(df, col="origin", row="tons", hue='origin')
g.map(sns.kdeplot, "horsepower", "mpg")
plt.xlim(0, 250)
plt.ylim(0, 60)

# 概率分布直方图，并拟合 连续性概率密度曲线
g = sns.FacetGrid(df, col="origin", row="tons")
g.map(sns.distplot, "mpg", bins=np.linspace(0, 50, 11))
plt.xlim(0, 50)
plt.ylim(0, 0.25)

# 频率分布直方图
g=sns.FacetGrid(df, col="origin", row="tons")
g.map(plt.hist, "mpg", bins=np.linspace(0, 50, 11))
plt.xlim(0, 50)
plt.ylim(0, 70)
# 同上，但要自己计算 分布的间隔
df['mpg_5']=(df.mpg/5).astype(int)
sns.factorplot("mpg_5", data=df, col="origin", row="tons", kind='count')
plt.xlim(0, 10)
plt.ylim(0, 70)


##pairplot
# 多个维度 ["mpg", "horsepower", "weight"] 两两组合绘图
g = sns.pairplot(df[["mpg", "horsepower", "weight", "origin"]], hue="origin",
                 kind='scatter', diag_kind="hist", size=5)
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)
    

##PairGrid
# 组合绘图，并回归
g = sns.PairGrid(df[["mpg", "horsepower", "weight", "origin"]], hue="origin")
g.map_upper(sns.regplot)
g.map_lower(sns.residplot)
g.map_diag(plt.hist)
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)
g.add_legend()
g.set(alpha=0.01)


##jointplot
# 多个变量联合绘图，一次性搞定
sns.jointplot("mpg", "horsepower", data=df, kind='kde')
# 拆分看看
sns.FacetGrid(df, size=1.5, aspect=3.5).map(sns.distplot, "mpg")
plt.xlim(0, 50)
plt.ylim(0, 0.05)
sns.FacetGrid(df, size=1.5, aspect=3.5).map(sns.distplot, "horsepower")
plt.xlim(0, 250)
plt.ylim(0, 0.0175)
sns.FacetGrid(df, size=5).map(sns.kdeplot, "mpg", "horsepower")
plt.xlim(0, 50)
plt.ylim(0, 250)
# 加回归参数
sns.jointplot("mpg", "horsepower", data=df, kind='reg', order=2)


##JointGrid
# 联合绘图：自定义绘图参数
g = sns.JointGrid(x="mpg", y="horsepower", data=df)
g.plot_joint(sns.regplot, order=1)
g.plot_marginals(sns.distplot)
```

```python
from pyspark import SparkContext, SQLContext
from pyspark.sql.functions import split
import os

def getCount(rdd, idx):
    idRDD = rdd.map(lambda x: x.split(',')[idx])
    count = len(set(idRDD.collect()))
    return count

def getAvgScore(rdd, idIdx, scoreIdx):
    scoreRDD = rdd.map(lambda x: (x.split(',')[idIdx], float(x.split(',')[scoreIdx])))
    avgScoreRDD = scoreRDD.groupByKey() \
                          .map(lambda x: (x[0], list(x[1]))) \
                          .map(lambda x: (x[0], round(sum(x[1]) / len(x[1]), 2)))
    # userScoreGroups = userScore.reduceByKey(lambda x, y: int(x) + int(y))
    return avgScoreRDD.collect()

def loadDf(filePath):
    rowDf = sqlContext.read.text(filePath)
    splitColumns = split(rowDf[u"value"], ",")
    rowDf = rowDf.withColumn(u"userId", splitColumns.getItem(0))
    rowDf = rowDf.withColumn(u"itemId", splitColumns.getItem(1))
    rowDf = rowDf.withColumn(u"score", splitColumns.getItem(2))
    rowDf = rowDf.drop(u"value")
    return rowDf

def getCountDF(df, colName):
    ids = df.select(colName).distinct()
    count = len(ids.collect())
    return count

def getAvgScoreDf(df, idCol, scoreCol):
    scoreDf = filteredDf.selectExpr(idCol, "float({})".format(scoreCol))
    avgScoreDf = scoreDf.groupby(idCol).mean()
    return avgScoreDf.show(3)
    
# 使用RDD处理数据
# sc = SparkContext()
# sc = SparkContext(master='master', appName='t1')
filePath = os.path.join(os.getcwd(), u"rating.txt")
rowRDD = sc.textFile(filePath)
filteredRDD = rowRDD.filter(lambda x: int(x.split(',')[2]) > 3)
print u"User count: ", getCount(filteredRDD, 0)
print u"Item count: ", getCount(filteredRDD, 1)
print u"User average score:\n", getAvgScore(filteredRDD, 0, 2)
print u"Item average score:\n", getAvgScore(filteredRDD, 1, 2)

# 使用DataFrame处理数据
# sc = SparkContext()
sqlContext = SQLContext(sc)
filePath = os.path.join(os.getcwd(), u"rating.txt")
rowDf = loadDf(filePath)
filteredDf = rowDf.filter("score > 3" )
print u"User count: ", getCountDF(filteredDf, u"userId")
print u"Item count: ", getCountDF(filteredDf, u"itemId")
print u"User average score:\n", getAvgScoreDf(filteredDf, u"userId", u"score")
print u"Item average score:\n", getAvgScoreDf(filteredDf, u"itemId", u"score")
```

```python
%config ZMQInteractiveShell.ast_node_interactivity='all'

from math import sqrt
import pandas as pd

def getPearsonDistance(rating1, rating2):
    sum_xy = 0
    sum_x = 0
    sum_y = 0
    sum_x2 = 0
    sum_y2 = 0
    n = 0
    for key in rating1:
        if key in rating2:
            n += 1
            x = rating1[key]
            y = rating2[key]
            sum_xy += x * y
            sum_x += x
            sum_y += y
            sum_x2 += pow(x, 2)
            sum_y2 += pow(y, 2)
    if n == 0:
        return 0
    denominator = sqrt(sum_x2 - pow(sum_x, 2) / n) * sqrt(sum_y2 - pow(sum_y, 2) / n)
    if denominator == 0:
        return 0
    else:
        score = (sum_xy - (sum_x * sum_y) / n) / denominator
        return round(score, 2)

def getSimilarityScore(itemIds, ratings):
    distances = [[0 for col in itemIds] for row in itemIds]
    for idx, item in enumerate(itemIds):
        for idxSim, itemSim in enumerate(itemIds):
            if ((idxSim > idx) and (item != itemSim)):
                distance = getPearsonDistance(ratings[item], ratings[itemSim])
                distances[idx][idxSim] = distance
                distances[idxSim][idx] = distance
    df = pd.DataFrame(distances, columns=itemIds, index=itemIds)
    return df

def transDict(rawDict):
    userIds = rawDict.keys()
    itemList = []
    for ratings in rawDict.values():
        itemList.extend(ratings.keys())
    itemIds = set(itemList)
    items = {}
    for itemId in itemIds:
        items[itemId] = {}
        for userId in userIds:
            if itemId in rawDict[userId].keys():
                items[itemId][userId] = rawDict[userId][itemId]
    return itemIds, items
    

users = {u"小明": {"中国合伙人2": 5.0, "太平轮": 3.0, "荒野猎人": 4.5, "老炮儿": 5.0, "我的少女时代": 3.0, "肖洛特烦恼": 4.5, "海王": 5.0},
        u"小红":{"小时代4": 4.0, "荒野猎人": 3.0, "我的少女时代": 5.0, "肖洛特烦恼": 5.0, "海王": 3.0, "后会无期": 3.0},
        u"小阳": {"小时代4": 2.0, "中国合伙人2": 5.0, "我的少女时代": 3.0, "老炮儿": 5.0, "肖洛特烦恼": 4.5, "速度与激情7": 5.0},
        u"小四": {"小时代4": 5.0, "中国合伙人2": 3.0, "我的少女时代": 4.0, "匆匆那年": 4.0, "速度与激情7": 3.5, "海王": 3.5, "后会无期": 4.5},
        u"六爷": {"小时代4": 2.0, "中国合伙人2": 4.0, "荒野猎人": 4.5, "老炮儿": 5.0, "我的少女时代": 2.0},
        u"小李":  {"荒野猎人": 5.0, "盗梦空间": 5.0, "我的少女时代": 3.0, "速度与激情7": 5.0, "蚁人": 4.5, "老炮儿": 4.0, "后会无期": 3.5},
        u"隔壁老王": {"荒野猎人": 5.0, "中国合伙人2": 4.0, "我的少女时代": 1.0, "Phoenix": 5.0, "甄嬛传": 4.0, "The Strokes": 5.0},
        u"邻村小芳": {"小时代4": 4.0, "我的少女时代": 4.5, "匆匆那年": 4.5, "甄嬛传": 2.5, "The Strokes": 3.0}
        }
userIds = users.keys()
getSimilarityScore(userIds, users)

itemIds, items = transDict(users)
getSimilarityScore(itemIds, items)
```
