```python
# 例子：如果array=[1, 1, 3, 1, 4, 2]，如果k=1，则返回3，有(3,4),(1,2),(2,3)这3对；如果k=2，则返回2，有(1,3),(2,4)这两对
def find_pairs(array, k):
    if len(array)<2 or k < 0:
        return 0
    array.sort()
    lower = 0
    higher = 1
    v_lower = -1
    v_higher = -1
    res = 0
    while lower < len(array) and higher < len(array):
        tmp_res = array[higher] - array[lower]
        if lower < higher and tmp_res >= k:
            if (tmp_res == k and (array[lower] != v_lower or array[higher] != v_higher)):
                res += 1
                v_lower = array[lower]
                v_higher = array[higher]
            lower += 1
        else:
            higher += 1
    return res

array=[1, 1, 3, 1, 4, 2]
k=2
find_pairs(array, k)
```
